```python
import nltk
import re

Words = nltk.corpus.words.words('en')
```

```python
cfd = nltk.ConditionalFreqDist((c,v) for w in Words
                              for (c,v) in re.findall('([kprstv])([aeiou])',w.lower()))
```

```python
cfd.tabulate()
```

```python
cfd.plot()
```

```python
cvs = [cv.lower() for w in Words
      for cv in re.findall('[ptksvr][aeiou]',w,re.I)]
cfd = nltk.ConditionalFreqDist(cvs)
```

```python
cfd.tabulate()
```

```python
cfd.plot()
```
