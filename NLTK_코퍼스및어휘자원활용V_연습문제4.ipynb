```python
import nltk
```

```python
names = nltk.corpus.names
```

```python
male = names.words('male.txt')
female = names.words('female.txt')
```

```python
Counts = nltk.ConditionalFreqDist(
            (fileid,name[-1])
            for fileid in names.fileids()
            for name in names.words(fileid))
```
