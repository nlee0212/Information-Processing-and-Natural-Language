```python
from konlpy.tag import Kkma
import os
```

```python
kkma = Kkma()
```

```python
file_list = ['./15_data/'+f for f in os.listdir('./15_data')]
```

```python
tagged = []
for f in file_list:
    tagged+= (kkma.pos(open(f,encoding='utf-8').read()))
```

```python
tagged
```

```python
len(tagged)
```

```python
print(tagged[:10])
```

```python
print(tagged[-10:])
```

# 모범답안

```python
a = os.getcwd()
b = os.listdir('./15_data')
tagged = [j for i in b for j in kkma.pos(open('/'.join([a,'15_data',i]),encoding='utf-8').read())]
```

```python
len(tagged)
```

```python
print(tagged[:10])
```
